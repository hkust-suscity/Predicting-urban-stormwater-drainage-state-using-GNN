```python
import numpy as np
import os
import pandas as pd
from datetime import datetime, timedelta
```

```python
temp=pd.read_csv('A_TEMP_223132_1141704.csv',skiprows=3,usecols=[0,2])
temp=temp.groupby(['Time']).mean()
temp=temp.reset_index()
temp['Time'] = pd.to_datetime(temp.Time, format = '%Y/%m/%d %H:%M:%S')
temp['Date']=temp['Time'].dt.strftime('%Y-%m-%d')
```

```python
max=temp.groupby(['Date']).max()
min=temp.groupby(['Date']).min()
max,min
```

```python
climate=pd.concat([max['Degree Celsius'],min['Degree Celsius']],axis=1)
climate=climate.reset_index()
climate.columns =['Date','max','min']
```

```python
df = climate
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
start_date = '2018-07-01'
end_date = '2023-06-30'
date_range = pd.date_range(start_date, end_date, freq='D')
new_df = pd.DataFrame(index=date_range)
df = df.interpolate(method='linear')
missing_values = new_df.loc[~new_df.index.isin(df.index)]
print(missing_values)
```

```python
df.loc['2023-01-16', 'max'] = np.nan
df.loc['2023-01-16', 'min'] = np.nan
df.loc['2023-01-17', 'max'] = np.nan
df.loc['2023-01-17', 'min'] = np.nan
df = df.sort_index(ascending=True)
df = df.interpolate(method='linear')
missing_values = new_df.loc[~new_df.index.isin(df.index)]
print(missing_values)
```

```python
climate = df
climate = climate.reset_index()
climate['year']=pd.DatetimeIndex(climate['Date']).year
climate['month']=pd.DatetimeIndex(climate['Date']).month
climate['day']=pd.DatetimeIndex(climate['Date']).day
climate['station']='KP_AWS'
climate['eva']='*'
climate
```

```python
wind = pd.read_csv('A_WIND_223132_1141704.csv',skiprows=3,usecols=[0,2])
wind = wind.groupby(['Time']).mean(numeric_only=True)
wind = wind.reset_index()
wind['Time'] = pd.to_datetime(wind.Time, format = '%Y/%m/%d %H:%M:%S')
wind['Date'] = wind['Time'].dt.strftime('%Y-%m-%d')
wind['km/h'] = wind['m/s']*60*60/1000
wind_average = wind.groupby(['Date']).mean(numeric_only=True)
wind_average = wind_average.reset_index()
wind_average
```

```python
climate=pd.concat([climate,wind_average['km/h']],axis=1)
climate
```

```python
climate.columns
```

```python
climate=climate.reindex(columns=['station', 'year', 'month', 'day','max', 'min', 'eva', 'km/h'])
climate
```

```python
climate=climate.round(1)
climate
```

```python
climate.to_csv('climate2018_2023.dat',sep='\t',header=False, index=False)
```
